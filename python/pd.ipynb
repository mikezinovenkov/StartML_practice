```python
data = df.drop_duplicates(subset=['App'])
```

```python
data.reset_index(drop= True)
```

```python
data
```

```python
data.columns = [x.replace(' ','_').lower() for x in data.columns]
```

```python
data
```

```python
data[data['type'] == 'Free']['type'].count()/data['type'].count()
```

```python
data
```

```python
data[(data['category'] == 'EDUCATION') & (data['reviews'] > 1000)].reset_index(drop = True)
```

```python
data['price'] = [float(x.replace('$','')) for x in data['price']]
```

```python
data.sort_values(by='price')
```

```python
ans = data.groupby(['category','type']).agg({'price':'mean', 'rating':'mean', 'reviews':'mean'})\
    .rename(columns={'price':'mean_price','rating':'mean_rating','reviews':'mean_reviews'})
```

```python
ans['mean_price'] = [round(x,2) for x in ans['mean_price']]
ans['mean_rating'] = [round(x,1) for x in ans['mean_rating']]
ans['mean_reviews'] = [round(x,2) for x in ans['mean_reviews']]
```

```python
ans
```

```python
svod = data.groupby(['category','type']).agg(mean_price=('price','mean'), mean_rating=('rating','mean'),\
                                       mean_reviews=('reviews','mean'))
svod.mean_rating = svod.mean_rating.round(1)
svod.mean_price = svod.mean_price.round(2)
svod.mean_reviews = svod.mean_reviews.round(2)

svod.to_csv('svod.csv', sep=',')
```

```python
new = pd.pivot_table(data, index=["category", "type"], aggfunc='mean').rename(columns={'price':'mean_price','rating':'mean_rating','reviews':'mean_reviews'})
```

```python
new['mean_price'] = [round(x,2) for x in new['mean_price']]
new['mean_rating'] = [round(x,1) for x in new['mean_rating']]
new['mean_reviews'] = [round(x,2) for x in new['mean_reviews']]
```

```python
new
```

```python
new.to_csv('new.csv', sep=',')
```
